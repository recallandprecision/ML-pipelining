## Null Imputation 

```python
df1['Outlet_Size'] = df1['Outlet_Size'].fillna(df1.Outlet_Size.mode()[0])
df2['Outlet_Size'] = df2['Outlet_Size'].fillna(df1.Outlet_Size.mode()[0])
```

##### Item weight is Nan Not at random so dropping

```python
df1=df1.dropna()
df2=df2.dropna()
```

```python
df1.shape
```

```python
data=df1.copy()
```

### converting string type to categorical and rest to float

```python

for i in data:
    if type(data[i][0])==float or type(data[i][0])==int:
        data[i]=data[i].astype(float)
    else:
        data[i]=data[i].astype('category').cat.codes
```

```python
data.dtypes
```

```python
data.head()
```

#### Quasi Constant handling with threshold=0.1

```python
sel = VarianceThreshold(0.1)  # 0.1 indicates 99% of observations approximately

sel.fit(data)
(sel.get_support())
print(
    len([
        x for x in data.columns
        if x not in data.columns[sel.get_support()]
    ]))

quasi_drop=[x for x in data.columns if x not in data.columns[sel.get_support()]]
print(quasi_drop)
data=data.drop(quasi_drop,1)
```

```python
data.head()
```

##### Correlation check 

```python
corr_matrix = data.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]
to_drop
```

```python
data.corr()
```

```python
corrmat = data.corr()
fig, ax = plt.subplots()
fig.set_size_inches(11,11)
sns.heatmap(corrmat,annot=True)
```

##### Dropping Correlated variable

```python
data=data.drop(to_drop,1)
```

```python
corrmat = data.corr()
fig, ax = plt.subplots()
fig.set_size_inches(11,11)
sns.heatmap(corrmat,annot=True)
```

##### no correlation exists now
